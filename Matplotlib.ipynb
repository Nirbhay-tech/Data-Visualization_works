```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

**Matplotlib Basics**

```python
x=np.linspace(0,5,11)
x
y=x**3
y
```

```python
plt.plot(x,y)
plt.title("TBSM")
plt.xlabel("Chachi")
plt.ylabel("Chacha")
```

**Subplots in Matplotlib**

```python
plt.subplot(2,2,1)
plt.plot(x,y)
plt.subplot(2,2,2)
plt.plot(y,x)
plt.subplot(2,2,3)
plt.plot(x**2,x**2)
plt.subplot(2,2,4)
plt.plot(x,y**0.5)
```

**Object Oriented Way**

```python
fig = plt.figure(dpi=(150))
axis1=fig.add_axes([0.1,0.1,.8,.8])
axis1.plot(x,y)
axis1=fig.add_axes([0.4,0.4,.2,.3])
axis1.plot(x,y)
```

**Types of Plots**

```python
plt.scatter(x,y)
```

```python
from random import sample
data = sample(range(1,1000),100)
plt.hist(data)
```

```python
plt.boxplot(x)
```

**Saving Plots**

```python
fig=plt.figure()
ax=fig.add_axes([.1,.1,.7,.7])
ax.plot(x,y)
plt.savefig('basicplot.png')
```
